```python
#Replace all null values for Categorical value by mode.
df_cd.fillna(df_cd.mode(axis=0,numeric_only=False).iloc[0,:])
```

```python
df_cpd.fillna(df_cpd.mode(axis=0,numeric_only=False).iloc[0,:])
```

```python
#Outlire
q1=df_cpd.quantile(0.25)
q3=df_cpd.quantile(0.75)
iqr=q3-q1
l=q1-(1.5*iqr)
u=q3+(1.5*iqr)
((df_cpd<l)|(df_cpd>u)).sum()
```

#### replace all the outliers values for numeric values by mean

```python
df_cpd['Annual Premium (in Rs)']=df_cpd['Annual Premium (in Rs)'].astype('float')
q1=df_cpd[['Annual Premium (in Rs)']].quantile(0.25)
q3=df_cpd[['Annual Premium (in Rs)']].quantile(0.75)
iqr=q3-q1
l=q1-(1.5*iqr)
u=q3+(1.5*iqr)
((df_cpd[['Annual Premium (in Rs)']]<l)|(df_cpd[['Annual Premium (in Rs)']]>u)).sum()
```

```python

correaltion_coefficient=np.corrcoef(master_table["Age"],master_table["Annual Premium (in Rs)"])
print("Correlation coefficient between Person Age and annual premium is {:.4f}".format(correaltion_coefficient[1][0]))
if correaltion_coefficient[1][0]>0.5:
    print("Strong positive relationship")
elif correaltion_coefficient[1][0]<-0.5:
    print("Strong negative relationship")
else:
    print("There is no relationship")
     
```
